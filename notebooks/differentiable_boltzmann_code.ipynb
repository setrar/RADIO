# Differentiable Boltzmann Code

```julia
using Bolt, Plots

# Assign cosmological parameters
 = CosmoParams(_c = 0.3) # set kwargs like so to change the default values
# Compute expansion history quantities
bg = Background()
# Compute ionization history (via RECFAST)
 = Bolt.RECFAST(bg=bg, Yp=.Y_p, OmegaB=._b)
ih = IonizationHistory(, , bg)

# Matter power spectrum
kmin,kmax,nk = 10bg.H,5000bg.H,32
ks = log10_k(kmin,kmax,nk) # k grid
pL = [plin(k,,bg,ih) for k in ks]
p1 = plot(ks, vcat(pL...), xscale=:log10, yscale=:log10)


# CMB C()
min,max,n = 2,20,1200
s = min:max:n
kmin,kmax,nk = 0.1bg.H, 1000bg.H, 100
ks = quadratic_k(kmin,kmax,nk)
sf = source_grid(, bg, ih, ks, BasicNewtonian()) # set up LOS source function interpolator
C = cltt(s, , bg, ih, sf)
p2 = plot(s, @.(s^2*C))
```
