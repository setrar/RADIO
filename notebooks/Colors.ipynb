```julia
using Plots
using Colors

# Define normalized RGB color
normalized_color = RGB(0.4, 0.4, 0.4)

# Convert to standard 8-bit RGB values
red = round(Int, red(normalized_color) * 255)
green = round(Int, green(normalized_color) * 255)
blue = round(Int, blue(normalized_color) * 255)

println("RGB Values: (", red, ", ", green, ", ", blue, ")")

# Use normalized RGB color in plot
plot_color = RGB(red/255, green/255, blue/255)

# Example plot using the converted RGB color
x = 1:10
y = rand(10)

plot(x, y
    , linewidth=2
    , linecolor=plot_color
    , label="Example Line"
    , xlabel="X-axis", ylabel="Y-axis"
    , title="Example Plot with Custom RGB Color"
)
```
